```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\DELL\Downloads\IBM_CHURN_DATA\telco.csv')
df.head()
df.info()
print(df.shape)
print(df.columns)
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Offer','Churn Category','Churn Reason'],inplace = True)
```

```python
df.info()
```

```python
df['Internet Type'].fillna(df['Internet Type'].mode()[0], inplace = True)
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip() #strip spaces in column names
df['Total Charges'] = pd.to_numeric(df['Total Charges'],errors = 'coerce')
df.isnull().sum()
```

```python
#churn by contract type
sns.countplot(x='Contract',hue= 'Churn Label', data=df)
plt.title('churn by contract type')
plt.show()
```

```python
#EDA
#Churn count by seaborn
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Churn Label',data=df)
plt.title('Churn Distribution')
plt.show()
```

```python
#churn by internet type
sns.countplot(x= 'Internet Type', hue = 'Churn Label', data=df)
plt.title('churn by Internet Type')
plt.xticks(rotation=45)
plt.show()
```

```python
#tenure vs churn
sns.histplot(data= df, x= 'Tenure in Months', hue= 'Churn Label', bins=30)
plt.title('Tenure vs churn')
plt.show()
```

```python
#monthly charges vs churn
sns.boxplot(x= 'Churn Label',y = 'Monthly Charge', data = df)
plt.title('Monthly charges and churn')
plt.show()
```

```python
gender_churn = df.groupby('Gender')['Churn Label'].value_counts(normalize = True).unstack()
gender_churn = gender_churn *100
print(gender_churn)
```

```python
sns.countplot(x= 'Gender', hue = 'Churn Label',data=df)
plt.title('churn by gender')
plt.show()
```

```python
#Analysis by matplotlib
#Univariate Analysis
#Churn Distribution
df['Churn Label'].value_counts().plot(kind='pie',autopct='%1.1f%%', colors=['lightgreen','salmon'])
plt.title('churn Distribution')
plt.ylabel("")
plt.show()
```

```python
#Gender vs churn Distribution
sns.countplot(x='Gender',hue='Churn Label',data=df)
```
